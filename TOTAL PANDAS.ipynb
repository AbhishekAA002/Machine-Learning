6. PANDAS - CLEANING EMPTY CELLS

NOTE : - EMPTY CELLS CAN POTENTIALLY GIVES WRONG RESULTS WHEN YOU ANALYZE THE DATA

Remove Rows



```python
#TO REMOVE ROWS THAT CONTAIN EMPTY CELLS
#df = pd.read_csv()
import pandas as pd
data = {
'Name':['Abhi','DON','PANDA','CAT','DOG',None,None],
'Age':[19,22,22,34,35,39,40]
    }
df = pd.DataFrame(data)
print('Data Frame with Contains empty cells \n',df)

ndf = df.dropna()
print('New Data Frame withnon empty rows : \n',ndf)
```

7.) REPLACE EMPTY VALUES

Another way of dealing with empty cells is to insert a new value instead
The fillna() method allows us to relace empty cells with a value

```python
#implace =True will modify the current data frame istead of returning the new one
ddf = df.fillna(130)
ddf
```

PANDAS - FIXING WRONG DATA 

WRONG DATA :- This doesn't have empty cells or wrong format , it can just be wrong , like if someone registered 199 instead of 1.99



```python
data = {
    'Duration':[60,60,60,45,45,60,60,450],
    'Date':['2020/12/01','2020/12/02','2020/12/03','2020/12/04','2020/12/05','2020/12/06','2020/12/07','2020/12/01'],
    'Pulse':[110,119,103,109,117,102,110,104],
    'Maxpulse':[130,145,135,175,148,1127,136,134],
    'Calories':[409.1,479.0,340.0,282.0,406.0,300.0,374.0,253.3]
}
df = pd.DataFrame(data)
df
```

```python
#replacing the value 450 to 45 in 7th row
df.loc[7,'Duration'] = 45
df
```

```python
df.loc[4,'Duration']
```

```python
#if there is large data set 
for i in df.index:
  if df.loc[i,'Maxpulse'] >130:
    df.loc[i,'Maxpulse'] = 120

df
```

```python
for i in df.index:
  if df.loc[i,'Maxpulse'] == 120 :
    df.drop(i,inplace = True) 
df
```

```python
#changing the column name 
df.rename(columns= {'Maxpulse':'MAXPULSE'},inplace = True)
df
```
