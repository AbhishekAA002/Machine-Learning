MATPLOTLIB PLOTTING

PLOTTING X AND Y POINTS 
THE plot() function is used to draw a line from point to point ._______.

This function takes parameters for specifying points in the diagram.

parameter 1 is an array containg the points on x asis 

parameter 2 is an array containg the points on y axis 

if we need to plot a line from (1,3) to (8,10) , we have to pass two arrays [1,8] and [3,10] 
parameter 1: - [x1,x2,x3,x4....,xn]
parameter 2: - [y1,y2,y3,y4,.....,yn]

The mapping will done [(x1,y1),(x2,y2)...(xn,yn)]


```python
#Draw a line in a digram from position (1,3) to position (8,10)
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,8])
y = np.array([3,10])

plt.plot(x,y)
plt.show()
```

MATPLOTLIB LABELS AND TITLE 

CREATE LABELS FOR A PLOT

plt.label('' , loc = '') - title
plt.xlabel('',fontdict) - xaxis 
plt.ylabel('',fontdict) - yaxis

WITH PYPLOT YOU CAN USE xlabel() AND  ylabel() functions to set a label for x-axix and y-axis

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)

plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.show()
```

CREATE A TITLE FOR A PLOT

title()

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title('WATCH DATA ')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

SET FONT PROPERTIES FOR TITLES AND LABELS
(family,color,size)
fontdict

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'red','size':15}

plt.title('Sports watch Data ',fontdict = font1)
plt.xlabel('Average Pulse',fontdict = font2)
plt.ylabel('Calorie Burnage',fontdict = font2)
plt.show()
```

POSITION OF THE TITLE


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'red','size':15}

plt.title('Sports watch Data ',fontdict = font1,loc = 'right')
plt.xlabel('Average Pulse',fontdict = font2)
plt.ylabel('Calorie Burnage',fontdict = font2)
plt.show()
```

MATPLOTLIB ADDING GRID LINES IN BACKGROUND 

WITH pyplot, you can use the grid() function to add grid lines to plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid()
plt.show()
```

SET LINE PROPERTIES FOR GRID LINE 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y,color='black',linewidth=4)

plt.grid(color = 'red',linestyle = '--',linewidth = 2)

plt.show()
```
